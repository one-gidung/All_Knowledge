# 3장 집약

### 3.1 데이터와 종류의 개수 산출

###### 집계와 유니크 카운트

DataFrame에서 group_by 함수를 호출해 매개변수에 집약 단위를 설정한 후 집약함수를 호출해 구현

size() : 데이터 수를 계산하는 집약함수

numique() : 유니크 카운트를 실행하는 집약함수

agg() : 같은 집약 단위로 여러 집약 처리를 실행하는 함수

```python
##### Not Awesome

# groupby 함수로 hotel_id를 집약 단위로 지정하고 size 함수로 데이터 수를 카운팅한다.
# groupby 함수에 의해 행 번호(index)가 흩어지므로 hotel_id가 집약 단위인 상태에서 reset_index 함수로 새로운 행 이름을 현재 행 번호로 고친다.
rsv_cnt_tb = reserve_tb.groupby('hotel_id').size().reset_index()                # reset_index(drop=True) : 기존 인덱스 삭제(여기서 인덱스는 hotel_id)

# 집약 결과의 열 이름을 설정한다.
rsv_cnt_tb.columns = ['hotel_id', 'rsv_cnt']

# groupby로 hotel_id를 집약 단위로 설정하고,
# customer_id 값을 nunique 함수를 이용하여 고객 수를 카운팅한다.
cus_cnt_tb = reserve_tb.groupby('hotel_id')['customer_id'].nunique().reset_index()

# 집약 결과의 열 이름을 설정한다.
cus_cnt_tb.columns = ['hotel_id', 'cus_cnt']

# merge 함수를 이용해 hotel_id를 결합 키로 결합한다.
pd.merge(rsv_cnt_tb, cus_cnt_tb, on='hotel_id')
```

집약한 열에 이름을 붙이려면 집약 후에 columns를 설정

groupby 함수를 이용하면 행 번호 대신 groupby 함수에 지정된 열의 값별로 행이 나뉜다.

즉, 인덱스가 행 번호에서 'hotel_id'로 바뀐다.

이후 처리에서 원치 않는 동작을 할 수도 있으므로 집약 처리 후 필요에 따라 reset_index 함수로 집약 후의 행 번호로 바꿔두자.

```python
##### Awesome

# agg 함수를 이용하여 집약 처리를 한꺼번에 지정한다.
# reserve_id를 대상으로 count 함수를 적용한다.
# customer_id를 대상으로 nunique 함수를 적용한다.
result = reserve_tb.groupby('hotel_id').agg({'reserve_id' : 'count',
                                             'customer_id' : 'nunique'})        # agg 함수의 매개변수에 dictionary 오브젝트를 지정

# reset_index 함수로 열 번호를 다시 설정한다(inplace=True이므로 직접 result를 갱신한다.)
result.reset_index(inplace=True)
result.columns = ['hotel_id', 'rsv_cnt', 'cus_cnt']
```

### 3.3 최댓값, 최솟값, 대푯값 산출

###### Q1 대푯값

예약 테이블에서 호텔별 예약 금액의 최댓값, 최솟값, 평균값, 중앙값, 20백분위 값(근삿값도 가능)을 산출

pandas 라이브러리 : max(최댓값), min(최솟값), mean(평균), median(중앙값)

numpy 라이브러리 : percentile(백분위)

```python
##### Awesome

# max/min/mean/median 함수에 total_price를 대상으로 적용한다.
# 파이썬의 람다식을 agg 함수의 집약 처리에 지정한다.
# 람다식에는 numpy.percentile을 지정하여 백분위 값을 산출한다(20% 지정).
result = reserve_tb.groupby('hotel_id').agg({'total_price' : ['max', 'min', 'mean', 'median', lambda x : np.percentile(x, q=20)]}).reset_index()
result.columns = ['hotel_id', 'price_max', 'price_min', 'price_mean', 'price_median', 'price_20per']

# groupby로 재정렬 -> agg로 병합 -> reset_index로 다시 지정 -> 열 이름 부여
# agg 함수 : 한 행에 여러 함수 적용 가능 (딕셔너리 형으로 입력 받음)
```

분포 정도: 분산값과 표준편차값을 이용해 나타낸다.

이때 데이터 수가 1일 때 분산은 n-1로 나누어주게 되어 0으로 나누어 잘못된 값이 나온다.

###### Q1 분산값과 표준편차값

단, 예약이 1건이면 분산값과 표준편차값을 0으로 한다.

### 3.6 순위 계산

대상 데이터를 추리거나 복잡한시간 데이터를 결합할 때 시간 순서로 순위를 매겨서 결합의 조건으로 이용

        집약한 결과를 계산하여 각 행에 첨부한다.

####### 순위를 구하는 함수

min : 같은 값의 순위 2~5위를 최소인 2위로 매긴다.

max : 같은 값의 순위 2~5위를 최대인 5위로 매긴다.

random : 같은 값의 순위 2~5위에서 랜덤하게 순위를 매긴다(중복되지 않는다.)

average : 같은 값의 순위 2~5위의 평균값인 3.5로 순위를 매긴다.

###### Q1 시간 데이터에 번호 부여

예약 테이블을 이용하여 고객별 예약 시간에 따른 순위를 매겨보자.

```python
##### Awesome

# rank 함수로 정렬하기 위해서 자료형을 문자열에서 timestamp형으로 변환한다.
reserve_tb['reserve_datetime'] = pd.to_datetime(
    reserve_tb['reserve_datetime'], format='%Y-%m-%d %H:%M:%S'
)

# log_no를 새로운 열로 추가한다.
# group_by를 이용하여 집약 단위 설정한다.
# 고객별로 묶은 reserve_datetime을 생섯하여 rank 함수로 순위를 계산한다.
# ascending을 True로 하여 오름차순으로 정렬한다(False면 내림차순).
reserve_tb['log_no'] = reserve_tb \
    .groupby('customer_id')['reserve_datetime'] \
    .rank(ascending=True, method='first')
```

rank 함수는 문자열을 처리할 수 없다.

이번 예제에서는 기준이 예약 시간이므로 자료형을 문자열에서 timestamp형으로 변환해야 한다.

rank 함수에서 method 매개변수에 같은 값의 데이터가 존재할 경우 순위 정하는 방법을 지정한다.

###### Q2 랭킹

호텔별 예약 건수에 따른 순위

```python
##### Awesome

# 예약 횟수를 계산한다
rsv_cnt_tb = reserve_tb.groupby('hotel_id').size().reset_index()
rsv_cnt_tb.columns = ['hotel_id', 'rsv_cnt']

# 예약 회수를 기준으로 순위를 계산한다.
# ascending을 False로 하여 내림차순으로 지정한다.
# method를 min으로 지정하여, 값이 같은 경우엔 최소 순위를 지정한다.
rsv_cnt_tb['rsv_cnt_rank'] = rsv_cnt_tb['rsv_cnt'] \
    .rank(ascending=False, method='min')

# 불필요한 rsv_cnt 열을 제거한다.
rsv_cnt_tb.drop('rsv_cnt', axis=1, inplace=True)
```
