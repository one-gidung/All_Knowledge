조건에 따른 결합 처리 : 결합 키를 위한 열을 새로 생성 -> 필요한 공통 열로 하나의 테이블을 만듭니다.

###### 마스터 테이블 변경

```python
small_area_mst = hotel_tb \
    .groupby(['big_area_name', 'small_area_name'], as_index=False) \
    .size().reset_index(drop=True)
small_area_mst
```

```python
small_area_mst = hotel_tb \
    .groupby(['big_area_name', 'small_area_name'], as_index=True) \
    .size().reset_index()
small_area_mst
```

```python
###### Awesome

# 가비지 컬렉션(불필요한 메모리 해제)을 위한 라이브러리
import gc

# small_area_name별로 호텔 수를 계산한다.
small_area_mst = hotel_tb \
    .groupby(['big_area_name', 'small_area_name'], as_index=False) \
    .size().reset_index(drop=True)
small_area_mst.columns = ['big_area_name', 'small_area_name', 'hotel_cnt']

# 20건 이상이면 join_area_id를 small_area_name으로서 지정한다.
# 20건 미만이면 join_area_id를 big_are_name으로서 지정한다.
# -1로 자신을 제외한다.
small_area_mst['join_area_id'] = \
    np.where(small_area_mst['hotel_cnt'] - 1 >= 20,
            small_area_mst['small_area_name'],
            small_area_mst['big_area_name'])

# 필요 없어진 열을 제거한다.
small_area_mst.drop(['hotel_cnt', 'big_area_name'], axis=1, inplace=True)

# 추천 전 호텔에 xmall_area_mst를 결합하여 join_area_id를 설정한다.
base_hotel_mst = pd.merge(hotel_tb, small_area_mst, on='small_area_name') \
    .loc[:, ['hotel_id', 'join_area_id']]

# 아래는 필요에 따라 메모리를 해제한다(필수는 아니지만, 메모리에 여유가 없을 때 사용).
del small_area_mst
gc.collect()

# recommend_hotel_mst는 추천 후보 테이블
recommend_hotel_mst = pd.concat([
    # join_area_id를 big_area_name으로 한 추천 후보 마스터
    hotel_tb[['small_area_name', 'hotel_id']] \
        .rename(columns={'small_area_name': 'join_area_id'}, inplace=False),
    
    # join_area_id를 small_area_name으로 한 추천 후보 마스터
    hotel_tb[['big_area_name', 'hotel_id']] \
        .rename(columns={'big_area_name': 'join_area_id'}, inplace=False)
])

# 결합할 때 hotel_id 열 이름이 중복되므로 변경한다.
recommend_hotel_mst.rename(columns={'hotel_id': 'rec_hotel_id'}, inplace=True)

# base_hotel_mst와 recommend_hotel_mst를 결합하여 추천 후보의 정보를 부여한다.
# query 함수로 추천 후보에서 자신을 제외한다.
pd.merge(base_hotel_mst, recommend_hotel_mst, on='join_area_id') \
    .loc[:, ['hotel_id', 'rec_hotel_id']] \
    .query('hotel_id != rec_hotel_id')
```

### 과거 데이터에서 정보 얻기

###### Q1 n건 전의 데이터 얻기

```python
###### Awesome

# customer별로 reserve_datetime으로 정렬한다.
# groupby함수 뒤에 apply 함수를 적용하여 group별로 정렬한다.
# sort_values 함수로 데이터를 정렬하고, axis가 0이면 행을, 1이면 열을 정렬한다.
result = reserve_tb \
    .groupby('customer_id') \
    .apply(lambda group:
          group.sort_values(by='reserve_datetime', axis=0, inplace=False))

# result는 이미 customer_id별로 그룹화 되었다.
# customer별로 두 건 이전의 total_price를 bfore_price로 저장한다.
# shift 함수는 periods 매개변수의 수만큼 데이터를 밑으로 옮기는 함수이다.
result['before_price'] = \
    result['total_price'].groupby('customer_id').shift(periods=2)
```

###### Q2 과거 n건의 합계값

```python
##### Awesome

# customer_id별로 reserve_datetime으로 데이터를 정렬한다.
result = reserve_tb.groupby('customer_id') \
    .apply(lambda x: x.sort_values(by='reserve_datetime', ascending=True)) \
    .reset_index(drop=True)

# 새로운 열로 price_sum을 추가한다.
result['price_sum'] = pd.Series(
    # 필요한 데이터열만 선택한다.
    result.loc[:, ["customer_id", "total_price"]]
    
    # cusvomer_id별로 total_price의 window3건의 합곗값을 계산한다.
    .groupby('customer_id')
    .rolling(center=False, window=3, min_periods=3).sum()
    
    # group화를 해제하면서 total_price의 열을 구한다.
    .reset_index(drop=True)
    .loc[:, 'total_price']
)
```

###### Q3 과거 n건의 평균값

rolling 함수는 자신의 데이터 행을 기준으로 한 선택만 할 수 있습니다.

그래서 이 문제를 처리하기 위해서 데이터 행을 옮길 수 있는 shift 함수를 사용

```python
##### Not Awesome

# customer_id별로 reserve_datetime으로 데이터를 정렬한다.
result = reserve_tb.groupby('customer_id') \
    .apply(lambda x: x.sort_values(by='reserve_datetime', ascending=True)) \
    .reset_index(drop=True)

# 새로운 열로 price_avg를 추가한다.
result['price_avg'] = pd.Series(
    result
        # customer_id별로 total_price의 sindow3건의 평균값을 계산한다.
        # min_periods를 1로 설정하여 1건 이상이면 계산하도록 설정한다.
        .groupby('customer_id')
        ['total_price'].rolling(center=False, window=3, min_periods=1).mean()
    
        # group화를 해제하면서 customer_id 열을 제거한다.
        .reset_index(drop=True)
)

# customer_id별로 price_avg를 1행 아래로 이동한다.
result['price_avg'] = \
    result.groupby('customer_id')['price_avg'].shift(periods=1)
```

###### Q4 과거 n일의 합곗값

```python
##### Not Awesome

import pandas.tseries.offsets as offsets
import operator

# 시간 계산에 사용하기 위해 자료형을 문자형에서 일시형으로 변환한다.
# (10장에서 자세히 설명한다.)
reserve_tb['reserve_datetime'] = \
    pd.to_datetime(reserve_tb['reserve_datetime'], format='%Y-%m-%d %H:%M:%S')

# reserve_datetime의 날짜를 확인하지 않고 customer_id가 같은 데이터들을 모두 결합한다.
sum_table = pd.merge(
    reserve_tb[['reserve_id', 'customer_id', 'reserve_datetime']],
    reserve_tb[['customer_id', 'reserve_datetime', 'total_price']]
        .rename(columns={'reserve_datetime': 'reserve_datetime_before'}),
    on='customer_id')

# checkin의 날짜를 비교하여 90일 이내 데이터가 결합된 데이터만을 추출한다.
# operator의 and_ 함수를 사용하영 결합 조건을 지정한다.
# reserve_id별로 total_price의 합을 계산한다.
# (날짜 데이터 형식은 10장에서 자세히 설명한다)
sum_table = sum_table[operator.and_(
    sum_table['reserve_datetime'] > sum_table['reserve_datetime_before'],
    sum_table['reserve_datetime'] + offsets.Day(-90) <= sum_table['reserve_datetime_before'])] \
    .groupby('reserve_id')['total_price'].sum().reset_index()

# 열 이름을 설정한다.
sum_table.columns = ['reserve_id', 'total_price_sum']

# 계산된 합을 결합하여 원본 테이블에 정보를 첨부한다.
# 합곗값이 없는 레코드의 합곗값을 fillna를 사용하여 0으로 변환한다.
pd.merge(reserve_tb, sum_table, on='reserve_id', how='left').fillna(0)
```

### 4.4 상호 결합

###### Q1 상호 결합 처리

파이썬에는 상호 결합을 위한 함수가 제공되지 않는다.

상호 결합을 구현하려면 모든 값이 같은 결합 키를 준비해야 한다.

```python
##### Awesome

# 날짜형을 위한 라이브러리
import datetime
# 날짜를 계산하기 위한 라이브러리
from dateutil.relativedelta import relativedelta

# 연월 마스터 테이블을 생성한다.
month_mst = pd.DataFrame({
    'year_month':
        # relativedelta로 2017-01-01을 x월간 보낸다. x에는 0, 1, 2를 대입한다.
        # 2017-01-01, 2017-02-01, 2017-03-01의 리스트를 생성한다.
        [(datetime.date(2017, 1, 1) + relativedelta(months=x)).strftime("%Y%m")
        for x in range(0, 3)]
})

# croos join을 위해 모두 같은 값을 가지는 결합 키를 준비한다.
customer_tb['join_key'] = 0
month_mst['join_key'] = 0

# customer_tb와 month_mst를 준비한 결합 키로 내부 결합하여 상호 결합을 구현한다.
customer_mst = pd.merge(
    customer_tb[['customer_id', 'join_key']], month_mst, on='join_key'
)

# 예약 테이블에서 연월의 결합 키를 준비한다.
reserve_tb['year_month'] = reserve_tb['checkin_date'] \
    .apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d').strftime("%Y%m"))

# 예약 레코드와 결합하여 사용 금액 함계를 계산한다.
summary_result = pd.merge(
    customer_mst,
    reserve_tb[['customer_id', 'year_month', 'total_price']],
    on=['customer_id', 'year_month'], how='left'
).groupby(['customer_id', 'year_month'])["total_price"] \
    .sum().reset_index()

# 예약 레코드가 없을 때 합계 금액을 값이 없음에서 0으로 변환한다.
summary_result.fillna(0, inplace=True)
```
